```python
#polymorphism
class Complex:
  def __init__(self,real,img):
    self.real=real
    self.img=img
  def showNumber(self):
    print(self.real,"i +",self.img,"j")

  def __add__(self, num2):#SUM OPERATOR OR DUNDER
    newReal= self.real + num2.real
    newImg= self.img + num2.img
    return Complex(newReal,newImg)

  def __sub__(self, num2):#sub dunder
    newReal= self.real - num2.real
    newImg= self.img - num2.img
    return Complex(newReal,newImg)


num1=Complex(5,6)
num1.showNumber()

num2=Complex(7,8)
num2.showNumber()

num3=num1+num2
num3.showNumber()

```

```python
#Define a Circle class to create with radius r using the constructior.
#Define an Area() method of the class which calculates the area of the circle
#Define a parimeter() method of thr class which allows you to calculate the perimeter of the circle.

class Cirle:
  def __init__(self,radius):
    self.radius=radius

  def area(self):
    return (22/7)*self.radius**2

  def perimeter(self):
    return 2*(22/7)*self.radius



c1=Cirle(21)
print(c1.area())
print(c1.perimeter())
```
